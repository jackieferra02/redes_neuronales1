```python
import numpy as np
import matplotlib.pyplot as plt

# Define la función sigmoide
def activation_function_sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define la función lineal para comparación
def activation_function_linear(x):
    return x

# Grafo computacional usando la función sigmoide
def computation_graph_sigmoid(x, w, b):
    y = activation_function_sigmoid(np.matmul(x, w) + b)
    return y

# Grafo computacional lineal
def computation_graph_linear(x, w, b):
    y = activation_function_linear(np.matmul(x, w) + b)
    return y

# Inicializa los valores del grafo computacional
def create_computation_graph_linear(n_in, n_out):
    w = np.random.randn(n_in, n_out) + 1
    b = np.random.randn(n_out, ) * 5
    return w, b

# Fijar semilla para controlar la aleatoriedad
np.random.seed(5)

# Número de puntos en el dominio para graficar las funciones
N_points_domain = 100

# Define nuestros datos
x_data = np.array([0, 1.5, 2, 2.5, 3, 4, 5]).reshape(7, 1)
t_data = np.array([0, 0, 0, 0, 1, 1, 1]).reshape(7, 1)

# Mostrar los datos nuevamente
plt.plot(x_data, t_data, 'o', markersize=8, label='data observations')
plt.xlabel('pixel')
plt.ylabel('etiqueta de clase')

# Especificación de la red neuronal para cada capa
n_in = 1
n_out = 1

# Crear varias funciones posibles que nuestra red neuronal específica puede implementar
for i in range(10):
    x_range = np.linspace(-1, 6, N_points_domain).reshape((N_points_domain, 1))

    # Inicializar una de nuestras redes
    w, b = create_computation_graph_linear(n_in, n_out)

    # Proyección de la entrada x a la salida y a través del grafo computacional usando sigmoide
    y_range = computation_graph_sigmoid(x_range, w, b)

    # Verificar cómo este grafo computacional predice en las entradas denotadas por nuestros datos observados X
    y_pred = computation_graph_sigmoid(x_data, w, b)

    if i == 0:
        plt.plot(x_data, y_pred, '*', markersize=5, color=f"C{i+1}", label='Network predictions at training input data')
        plt.plot(x_range, y_range, color=f"C{i+1}", label='NeuralNet implemented function on all the domain')
    else:
        plt.plot(x_data, y_pred, '*', markersize=5, color=f"C{i+1}")
        plt.plot(x_range, y_range, color=f"C{i+1}")

plt.legend()
plt.show()
```
