```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#pintar las derivadas de la funci√≥n y 

N_points_domain = 100

x_range = np.linspace (-2.5, 2.5, N_points_domain)

def funcion_y (x): #entrada a mi red neuronal
    result = (x**2-4)**3 + 2
    return result

def derivada_y (x):
    result = 3*(x**2-4)**2*2*x #ojo 2x es igual a 2*x
    return result

#para pintar ambas derivadas se hace el plot con ambas funciones 
plt.plot(funcion_y(x_range))
plt.plot(derivada_y(x_range))
```
